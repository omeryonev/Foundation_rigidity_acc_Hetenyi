# Foundation Rigidity according to Hetenyi

```python
import handcalcs
import handcalcs.render
import math
from math import sqrt, pi
from handcalcs.decorator import handcalc
import forallpeople
forallpeople.environment('structural' , top_level = True)
```

**Inputs**

```python
%%render short
E = (10000*kPa) # Modulus of Elasticity of Soil
v = (0.3) # Possion ratio of soil
B = (2*m)
E_b = (210000*MPa)
H = (0.25*m)
I = (B * H**3 / 12)
L = (10*m)
```

**Outputs**

```python
%%render short
k = 0.65/(1-v**2) * ((E * B**4) / (E_b * I))**(1/12) * E / B
lamb = ((k * B) / (4 * E_b * I))**(1/4) * L
```

```python
if lamb < pi / 4:
    print(lamb, "<", "pi/4=",pi / 4, "--> Rigid foundation")
elif lamb >= pi / 4 and lamb <= pi:
    print(lamb, ">=", "pi/4 =",pi / 4, "--> Intermediate rigid foundation")
elif lamb > pi:
    print(lamb, ">","pi", pi, "--> Flexible foundation")
```
